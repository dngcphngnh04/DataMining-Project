```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Customer-survey-data.csv')
data
```

```python
data.isnull()
```

```python
data = data.drop(columns=['Unnamed: 0','id','Departure Delay in Minutes','Arrival Delay in Minutes'], axis = 1)
data
```

```python
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
data['Gender'] = pd.to_numeric(data['Gender'])
data['Gender'].value_counts()
```

```python
data['Customer Type'] = data['Customer Type'].map({'Loyal Customer': 0, 'disloyal Customer': 1})
data['Customer Type'] = pd.to_numeric(data['Customer Type'])
data['Customer Type'].value_counts()
```

```python
bin_edges = [0, 25, 40, 50, np.inf]
data['Age'] = pd.cut(data['Age'], bins=bin_edges, labels = [0,1,2,3], include_lowest = True)
data['Age'] = pd.to_numeric(data['Age'])
data['Age'].value_counts()
```

```python
data['Type of Travel'] = data['Type of Travel'].map({'Business travel': 0, 'Personal Travel': 1})
data['Type of Travel'] = pd.to_numeric(data['Type of Travel'])
data['Type of Travel'].value_counts()
```

```python
data['Class'] = data['Class'].map({'Business': 0, 'Eco Plus': 1, 'Eco': 1})
data['Class'] = pd.to_numeric(data['Class'])
data['Class'].value_counts()
```

```python
bin_edges = [0, 400, 900, 1700, np.inf]
data['Flight Distance'] = pd.cut(data['Flight Distance'], bins=bin_edges, labels=[0,1,2,3], include_lowest=True)
data['Flight Distance'] = pd.to_numeric(data['Flight Distance'])
data['Flight Distance'].value_counts()
```

```python
for col in data.columns:
  plt.hist(data[col])
  plt.title(col)
  plt.show()
```

```python
for col in data.columns:
  sns.boxplot(data[col])
  plt.title(col)
  plt.show()
```

```python
data['satisfaction'] = data['satisfaction'].map({'neutral or dissatisfied': 'Dissatisfied', 'satisfied': 'Satisfied'})
```
